```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series, Index
import pandas as pd
import census
import us

import settings
from itertools import islice
```

```python
c = census.Census(key=settings.CENSUS_KEY)
```

```python
def metro(variables = 'NAME'):
    '''Generator for all metropolitan areas'''
    
    for state in us.states.STATES:
        geo = {'for': 'metropolitan statistical area/micropolitan statistical area:*', 'in': 'state:{fips}'.format(fips=state.fips)}
        for metro in c.sf1.get(variables, geo=geo):
            yield metro
        
```

```python
for item in df_comb.columns[2:]:
    df_comb['%'+item] = (df_comb[item] / df_comb.Total)
df_comb.head()
```

```python
def entropy(series):
    """Normalized Shannon Index"""
    # a series in which all the entries are equal should result in normalized entropy of 1.0
    
    # eliminate 0s
    series1 = series[series!=0]

    # if len(series) < 2 (i.e., 0 or 1) then return 0
    
    if len(series) > 1:
        # calculate the maximum possible entropy for given length of input series
        max_s = -np.log(1.0/len(series))
    
        total = float(sum(series1))
        p = series1.astype('float')/float(total)
        return sum(-p*np.log(p))/max_s
    else:
        return 0.0
    

    
df_comb['entropy5'] = df_comb[['Asian','Black','Hispanic','White','Other']].apply(entropy,axis=1)
df_comb['entropy4'] = df_comb[['Asian','Black','Hispanic','White']].apply(entropy,axis=1)
```

```python
df_comb.sort_index(by='entropy5', ascending=False).head()
```

```python
# help from: https://www.broadinstitute.org/~jmeng/picower/jmeng/CPAT_CodingPotentialAsessTool/CPAT/CPAT-1.2.1/lib/cpmodule/mystat.py
def simpson(series):
    '''gini simpson index'''
    
    series1 = series[series!=0]
    
    simpson=0.0
    
    try:
        for i in series1:
            simpson = simpson + (i/sum(series1))**2
        return 1-simpson
    except: return 0
    
df_comb['gini_simpson'] = df_comb[['Asian','Black','Hispanic','White','Other']].apply(simpson,axis=1)
df_comb.head()
```

```python
df_comb.columns = [[u'Name', u'Total', u'White', u'Black', u'Asian',
                    u'Other', u'Hispanic', u'p_White', u'p_Black', u'p_Asian',
                    u'p_Other', u'p_Hispanic', u'entropy5', u'entropy4', u'gini_simpson']]
```

```python
def rice(series):
    """Normalized Rice Index"""
    # a series in which all the entries are equal should result in normalized entropy of 1.0
    
    # eliminate 0s
    series1 = series[series!=0]

    # if len(series) < 2 (i.e., 0 or 1) then return 0
    
    if len(series) > 1:
        # calculate the maximum possible entropy for given length of input series
        max_s = -np.log(1.0/len(series+1))
    
        total = float(sum(series1))
        p = series1 #.astype('float')/float(total)
        return sum(-p*np.log(p))/max_s
    else:
        return 0.0
    
#Houston to come out to "0.874" and that for NYC to be "0.872"
    
df_comb['entropy_rice'] = df_comb[['p_Asian','p_Black','p_Hispanic','p_White']].apply(rice,axis=1)
df_comb.sort_index(by='entropy_rice', ascending=False).head()
```
